<font color='#045FB4'>
- We apply onehot encoding technique on 'Sex' and 'SibSp' features<br>
- 'Sex' having only two unique category male and female   

```python
data = pd.read_csv('Titanic.csv',usecols=['Sex'])
data.head()
```

<h3 style="color:#B4045F;"> Applying Onehot Encoding on 'Sex' Feature

```python
pd.get_dummies(data['Sex']).head()
```

<font color='#045FB4'>

- Here we performed onehot encoding technique on Sex feature 
- It create a two new features female and male and arrenge in alphabetical order
- In 'Sex' feature wherever female is category they put 1 in new 'female' feature else 0.
- Similarly, wherever male is category they put 1 in new 'male' feature else 0.
- From this we can say that if 'male' feature's value is 1 then female feature's value is 0
- So we drop 1st new feature by using 'drop_first=True' parameter and rename column name with 'Sex'

```python
dt = pd.get_dummies(data['Sex'],drop_first=True,columns=['Sex']).head()
dt.rename(columns={'male':'Sex'},inplace=True)
dt.head()
```

 1 = Male & 0 = Female

<h2 style="color:#B4045F;"> Apply onehot encoding when feature having multiple categories

```python
data = pd.read_csv('mercedes.csv')
data.head()
```

<h3 style="color:#B4045F;">
 we use first 6 columns/features only

```python
data = pd.read_csv('mercedes.csv',usecols=['X0','X1','X2','X3','X4','X5','X6'])
data.head()
```

<h3 style="color:#B4045F;">Find how many unique values are present in each feature

```python
for i in data.columns:
    print(i,'having',+len(data[i].unique()),'unique values')
```

<h3 style="color:#B4045F;">
Find count of top 10 frequent category in 'X0' feature

```python
data['X0'].value_counts().head(10)
```

```python
list_X0_10 = data['X0'].value_counts().head(10).index
list_X0_10
```

<h3 style="color:#B4045F;">
Convet index to list

```python
list_X0_10 = list(list_X0_10)
list_X0_10
```

<h3 style="color:#B4045F;">
Now perform onehot encoding on only this 10 categories

```python
for category in list_X0_10:
    data[category]=np.where(data['X0']==category,1,0)
```

```python
list_X0_10.append('X0')
```

```python
data[list_X0_10].head(15)
```
